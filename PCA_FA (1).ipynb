```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Load the Excel dataset into a pandas DataFrame
excel_file = 'Book1.csv'  # Replace 'your_excel_file.xlsx' with the path to your Excel file
df = pd.read_csv(excel_file)
```

```python
# Assume that your features are in columns 'feature1', 'feature2', ..., 'featureN'
features = df[['B6', 'B7', 'B8', 'B9']]

# Standardize the features
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

# Perform PCA
pca = PCA(n_components=1)  # Set n_components to 1 for reducing features to one dimension
principal_components = pca.fit_transform(scaled_features)

# Create a DataFrame with the principal components
principal_df = pd.DataFrame(data=principal_components, columns=['PC1'])

# Concatenate the principal components DataFrame with any other relevant columns
# For example, if you want to keep the target variable 'target_column':
# principal_df = pd.concat([principal_df, df['target_column']], axis=1)

# Optionally, you can also access the explained variance ratio of the principal components
explained_variance_ratio = pca.explained_variance_ratio_
print("Explained Variance Ratio of Principal Component:", explained_variance_ratio[0])

# Now 'principal_df' contains the dataset with reduced dimensionality
# You can save it to a new Excel file if needed
principal_df.to_excel('principal_components.xlsx', index=False)  # Change the filename as per your preference
```

```python
# Compute eigenvectors and eigenvalues
eigenvectors = pca.components_  # Eigenvectors
eigenvalues = pca.explained_variance_  # Eigenvalues

# Print eigenvectors and eigenvalues
print("Eigenvectors:\n", eigenvectors)
print("Eigenvalues:", eigenvalues)
```

```python
import matplotlib.pyplot as plt
# Plot the PCA
plt.figure(figsize=(8, 6))
plt.scatter(principal_df['PC1'], [0] * len(principal_df), color='b', alpha=0.5)  # Plot PC1
plt.xlabel('Principal Component 1')
plt.ylabel('Arbitrary Y-axis')
plt.title('PCA Plot')
plt.grid(True)
plt.show()
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import pandas as pd
import matplotlib.pyplot as plt

# Assume that your features are in columns 'feature1', 'feature2', ..., 'featureN'
features = df[['B6', 'B7', 'B8', 'B9']]

# Standardize the features
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

# Perform PCA
pca = PCA(n_components=2)  # Set n_components to 2 for reducing features to two dimensions
principal_components = pca.fit_transform(scaled_features)

# Create a DataFrame with the principal components
principal_df = pd.DataFrame(data=principal_components, columns=['PC1', 'PC2'])

# Compute eigenvectors and eigenvalues
eigenvectors = pca.components_  # Eigenvectors
eigenvalues = pca.explained_variance_  # Eigenvalues

# Plot the PCA
plt.figure(figsize=(8, 6))
plt.scatter(principal_df['PC1'], principal_df['PC2'], color='b', alpha=0.5)  # Plot PC1 vs PC2
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA Plot')
plt.grid(True)
plt.show()

# Print eigenvectors and eigenvalues
print("Eigenvectors:\n", eigenvectors)
print("Eigenvalues:", eigenvalues)

# Optionally, you can also access the explained variance ratio of the principal components
explained_variance_ratio = pca.explained_variance_ratio_
print("Explained Variance Ratio of Principal Components:", explained_variance_ratio)

# Now 'principal_df' contains the dataset with reduced dimensionality
# You can save it to a new Excel file if needed
principal_df.to_excel('principal_components.xlsx', index=False)
```

```python
def parse_table(tableData):
    data={'Postal Code':[],'Borough':[],'Neighborhood':[]}
    for row in tableData:
        for td in row.find_all('td'):
            code=td.find('b').text
            borough_=td.find('span').text
            # skip if borough_ is Not assigned
            if borough_=='Not assigned':
                continue
            borough=borough_.split('(')[0]
            neighborhood=re.findall('\((.*?)\)', borough_)[0].replace('/',',')

            # append into the dict
            data['Postal Code'].append(code)
            data['Borough'].append(borough)
            data['Neighborhood'].append(neighborhood)
    return data




```
