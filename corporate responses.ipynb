```python
#importing library
import pandas as pd
```

```python
#we convert the excel document to a dataframe 
df = pd.read_excel("corporate.xlsx")
```

```python
#we are just logging to see how the file looks like after conversion
df
```

The are a number of methods that can be used but i will use the replace method

- I will demonstrate replacing each column one after the other
- I will demonstrate replacing it all at onces

```python
#replacing all the values with D, N, A with the numbers. This replaces it in every column with those values
df2 = df.replace({'D': 1, 'N': 2, 'A': 3})
```

```python
#this is the results
df2
```

```python
#This gives me all the column titles of the file.
df2.columns
```

```python
#i want to find the uniques values of this column. I will then assign the numerical values
df2['Highest Qualification'].unique()
```

```python
# i will replace each value with their corresponding numerical value
df3 = df2.replace({'Certificate': 1, 'Diploma': 2, 'Bachelor Degree': 3, 'Master’s Degree': 4, 'PHD': 5})
```

```python
#the results
df3
```

```python
print(df2['1.     Firm Type'].unique())
print(df2['Gender'].unique())
print(df2['Years of Service'].unique())
print(df2['Age'].unique())
```

```python
#this time, i don't want to do one column at a time, I am creating a dictionary for all the columns left

replacement_dict = {
    '1.     Firm Type': {'Small': 1, 'Medium': 2, 'Big': 3},
    'Gender': {'Male': 1, 'Female': 2},
    'Years of Service': {'Less than a year': 1, '1-5 years': 2, '6-10 years': 3, '11-15 years': 4, 'At least 16 years': 5},
        'Age': {'20-29': 1, '30-39': 2, '40-49': 3, '50-59': 4}

}

dfFinal = df3.replace(replacement_dict)
```

```python
dfFinal
```

```python
#saving the files as csv
dfFinal.to_csv("corporateResponse.csv",index=False)
dfFinal.to_excel("corporateResponse.xlsx",index=False)
```
